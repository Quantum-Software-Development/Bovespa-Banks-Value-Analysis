```python
# Executa a amostra estratificada igualitÃ¡ria por estado civil
amostra_est_igual=df.groupby(['estado_civil'], group_keys=False).apply(lambda x: x.sample(n=10,replace=False,random_state=2903))

amostra_est_igual
```

```python
#definir total desejado
N = 20

# Executa a amostra estratificada proporcional por estado civil
amostra_est_prop = df.groupby(['estado_civil'], group_keys=False).apply(
    lambda x: x.sample(int(np.rint(N * len(x) / len(df))), random_state=2903)  # CÃ¡lculo proporcional da amostra
).sample(frac=1, random_state=2903).reset_index(drop=True)  # Embaralhar a amostra

amostra_est_prop
```


# min(len(x), 10)
Garante que, se um estrato tiver menos de 10 elementos, ele nÃ£o tentarÃ¡ amostrar mais do que o disponÃ­vel. Isso evita erros ao tentar amostrar mais elementos do que existem em um estrato especÃ­fico.

# DescriÃ§Ã£o do CÃ³digo
Este cÃ³digo cria uma amostra estratificada igualitÃ¡ria por grau de instruÃ§Ã£o e regiÃ£o de procedÃªncia, considerando atÃ© 10 casos por estrato. A amostra resultante Ã© exibida e salva em um novo arquivo Excel.

# ConclusÃ£o
A amostragem estratificada Ã© uma tÃ©cnica poderosa para garantir que todas as subpopulaÃ§Ãµes (estratos) sejam representadas adequadamente na amostra. Neste caso, a amostra foi estratificada por grau de instruÃ§Ã£o e regiÃ£o de procedÃªncia, garantindo uma representaÃ§Ã£o equilibrada de cada estrato. A amostra resultante foi salva em um novo arquivo Excel para futuras anÃ¡lises.

### ðŸ‘‡ Para salvar a amostra estratificada em um novo arquivo Excel, vocÃª pode usar o mÃ©todo `to_excel` fornecido pelo pandas. Abaixo estÃ¡ o cÃ³digo completo, incluindo os passos para salvar a amostra:

```python
# Importar bibliotecas pandas e numpy
import pandas as pd
import numpy as np

# Ler o arquivo Excel em um DataFrame
df=pd.read_excel(Caminho)

# Criar uma amostra estratificada igualitÃ¡ria por grau de instruÃ§Ã£o e regiÃ£o de procedÃªncia
amostra_est_igual = df.groupby(['grau_instrucao', 'reg_proc'], group_keys=False).apply(lambda x: x.sample(n=min(len(x), 10), replace=False, random_state=2903)).reset_index(drop=True)

# Exibir a amostra resultante
print(amostra_est_igual)

# Salvar a amostra estratificada em um novo arquivo Excel
caminho_saida = '/Users/fabicampanari/Desktop/amostra_estratificada.xlsx'
amostra_est_igual.to_excel(caminho_saida, index=False)

print(f"A amostra estratificada foi salva em {caminho_saida}")
```
