```python
# Executa a amostra estratificada igualit√°ria por estado civil
amostra_est_igual=df.groupby(['estado_civil'], group_keys=False).apply(lambda x: x.sample(n=10,replace=False,random_state=2903))

amostra_est_igual
```

```python
#definir total desejado
N = 20

# Executa a amostra estratificada proporcional por estado civil
amostra_est_prop = df.groupby(['estado_civil'], group_keys=False).apply(
    lambda x: x.sample(int(np.rint(N * len(x) / len(df))), random_state=2903)  # C√°lculo proporcional da amostra
).sample(frac=1, random_state=2903).reset_index(drop=True)  # Embaralhar a amostra

amostra_est_prop
```


# min(len(x), 10)
Garante que, se um estrato tiver menos de 10 elementos, ele n√£o tentar√° amostrar mais do que o dispon√≠vel. Isso evita erros ao tentar amostrar mais elementos do que existem em um estrato espec√≠fico.

# reset_index(drop=True)
Reseta o √≠ndice do DataFrame resultante para evitar √≠ndices duplicados. Isso √© √∫til para manter a integridade dos dados e facilitar a manipula√ß√£o subsequente do DataFrame.

# Descri√ß√£o do C√≥digo
Este c√≥digo cria uma amostra estratificada igualit√°ria por grau de instru√ß√£o e regi√£o de proced√™ncia, considerando at√© 10 casos por estrato. A amostra resultante √© exibida e salva em um novo arquivo Excel.

# Conclus√£o
A amostragem estratificada √© uma t√©cnica poderosa para garantir que todas as subpopula√ß√µes (estratos) sejam representadas adequadamente na amostra. Neste caso, a amostra foi estratificada por grau de instru√ß√£o e regi√£o de proced√™ncia, garantindo uma representa√ß√£o equilibrada de cada estrato. A amostra resultante foi salva em um novo arquivo Excel para futuras an√°lises.

### üëá Para salvar a amostra estratificada em um novo arquivo Excel, voc√™ pode usar o m√©todo `to_excel` fornecido pelo pandas. Abaixo est√° o c√≥digo completo, incluindo os passos para salvar a amostra:

```python
# Importar bibliotecas pandas e numpy
import pandas as pd
import numpy as np

# Ler o arquivo Excel em um DataFrame
df=pd.read_excel(Caminho)

# Criar uma amostra estratificada igualit√°ria por grau de instru√ß√£o e regi√£o de proced√™ncia
amostra_est_igual = df.groupby(['grau_instrucao', 'reg_proc'], group_keys=False).apply(lambda x: x.sample(n=min(len(x), 10), replace=False, random_state=2903)).reset_index(drop=True)

# Exibir a amostra resultante
print(amostra_est_igual)

# Salvar a amostra estratificada em um novo arquivo Excel
caminho_saida = '/Users/fabicampanari/Desktop/amostra_estratificada.xlsx'
amostra_est_igual.to_excel(caminho_saida, index=False)

print(f"A amostra estratificada foi salva em {caminho_saida}")
```
